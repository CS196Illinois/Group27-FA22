```python
from sportsipy.ncaab.conferences import Conferences

conferences = Conferences()
# Prints a dictionary of the team abbrevation as a key and conference
# abbreviation as the value.
conferences_dict = conferences.conferences
```

```python
temp = list(conferences_dict['big-ten']['teams'].values())
big_ten_teams = []
for team in temp:
    team = team.replace(" ", "-")
    big_ten_teams.append(team)
big_ten_teams
```

```python
#Illinois 2021-2022 most recent season on API rn
from sportsipy.ncaab.roster import Roster
Roster(team="Illinois")
```

```python
from sportsipy.ncaab.teams import Team
Team("ILLINOIS", year=2022).dataframe
```

```python
from sportsipy.ncaab.teams import Team

big_ten_df = pd.DataFrame()
for school_year in ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']:
    for team in big_ten_teams:
        data = Team(team.upper(),year = school_year)
        temp_df = data.dataframe
        temp_df.insert(1, column="Year", value=school_year)
        big_ten_df = pd.concat([big_ten_df, temp_df])
    

```

```python
big_ten_df = big_ten_df.reset_index(drop=True)
```
